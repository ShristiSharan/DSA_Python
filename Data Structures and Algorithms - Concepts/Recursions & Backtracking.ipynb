* It consume less memory space as the n number stack is created  at the same time whereas iteration takes up n times.

* But also if there's no base case given and recursion works infinitely then error: STACK OVERFLOW occurs....means memory 

*  Due to stack implementation - first it creats stack till the base case then reaching that clears down the memory after returning.
                                        


```python
def printSum(i,n,sum):
    if(i==n):
        sum+=i
        print(sum)
    sum +=i
    printSum(i+1,n,sum)

def Print():
    print(printSum(1,5,0))
```

```python
#Standard question of recursion - TOWER OF HANOI
# Condition:
#       * Big disc cannot be placed above shorter one. 

# source(S)---helper(H)--destination(D)
def towerOfHanoi(n,src,helper,dest):
   c=str(n)
   if(n == 1):
      print("transfer disk"+ c +" from "+src+" to "+dest)
      return
   towerOfHanoi(n-1,src,dest,helper)
   print("transfer disk "+ c +" from "+src+" to "+dest)
   towerOfHanoi(n-1,helper,src,dest)

n=1
towerOfHanoi(n,"S","H","D")


#O(2^n-1) ~= O(2^n)..........
 
```

```python
def printPerm(str,permutation):
    if(len(str)==0):
        print("permutation")
        return
    for i in range(len(str)):
        currChar=str[i]
        newStr= (str.substring(0,i) + str.substring(i+1))
        printPerm(newStr,permutation+currChar)

str="abc"
printPerm(str,"")
```
